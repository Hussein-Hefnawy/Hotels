```python
import pandas as pd
import numpy as py 
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('hotels.csv')
```

<p style="background-color:#6A5ACD;font-family:Verdana, sans-serif;color:#FFFFFF;font-size:150%;text-align:center;border-radius:10px;padding:10px;"> ProblemHandle missing Data

```python
df.drop(["company","agent"],axis = 1,inplace = True)
df.dropna(inplace = True)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['reservation_status_date'] =pd.to_datetime(df['reservation_status_date'],format='mixed')
```

```python
df.describe()
```

```python
df = df[df['adr']<5000]
```

```python
df.head(10)
```

<p style="background-color:#6A5ACD;font-family:Verdana, sans-serif;color:#FFFFFF;font-size:150%;text-align:center;border-radius:10px;padding:10px;"> Data Analysis and visualizations

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)*100
cancelled_perc
```

```python
cancelled_perc = df['is_canceled'].value_counts(normalize =True)
print(cancelled_perc)
plt.figure(figsize=(5,4))
plt.title('Reservation ststus count')
plt.bar(['Not canceled','canceled'],df['is_canceled'].value_counts(),edgecolor='b',width=0.5)
```

 # 37% of clients had cancelled their reservations, and this has a noteworthy impact on the hotels revenue

```python
plt.figure(facecolor="#30d195")
sns.countplot(x='hotel' , hue='is_canceled' , data = df)
plt.title('Cancelation rates in City hotel and Resort hotel',fontweight="bold", size=20)
plt.legend(['Not Cancelled ', 'Cancelled'])
plt.show()
```

# reservations cancelled in City Hotel is bigger than Resort Hotel
# Also reservations in City Hotel Is bigger than Resort Hotel , 
## ( maybe cause resort hotels are more expensive than in cities)

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize= True)
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize= True)
```

```python
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
resort_hotel
```

```python
plt.figure(figsize=(20,8),facecolor='#30d195')
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=15)
plt.grid()
plt.show()
```

# on certain days, the average daily rate for a city hotel is less than that of a resort hotel and on other days

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(15,7),facecolor='#C38154')
sns.countplot(x='month' , hue='is_canceled', data=df,palette='Pastel1')
plt.title('Reservation status per month',fontweight="bold", size=20)
plt.ylabel('number of reservations',fontweight="bold", size=15)
plt.xlabel('month',fontweight="bold", size=15)
plt.legend(['not canceled', 'canceled'])
plt.grid()
plt.show() 
```

```python
plt.figure(figsize = (15,8), facecolor='#30d195' )
plt.title('ADR per Month', fontsize = 30)
data = df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()
sns.barplot(x='month', y='adr', data = data)
plt.xlabel('month',fontweight="bold", size=15)
plt.ylabel('ADR',fontweight="bold", size=15)
plt.show()
```

```python
data = df[df['is_canceled'] == 0].groupby('month')['adr'].sum().reset_index()
data
```

## This bar graph illustrates that cancellations are most frequent when prices are at their highest and least common when prices are at their lowest. Consequently, the price of accommodation appears to be the primary factor influencing cancellations.
